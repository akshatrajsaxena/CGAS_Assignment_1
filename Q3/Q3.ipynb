```python
import pandas as pd
import math
df = pd.read_csv('../Q1/Q1_b.csv')
```

Counting frequency of unique ingredients

```python
freq = {}
for item in df['Ingredient Name']:
  if item in freq:
    freq[item] += 1
  else:
    freq[item] = 1
```

Storing List in Dataframe

```python
arr_ingredient = []
arr_freq = []
for ing, freq in freq.items():
  arr_ingredient.append(ing)
  arr_freq.append(freq)
```

```python
dict = {'Ingredient Name': arr_ingredient, 'Frequency': arr_freq}
required_df = pd.DataFrame(dict)
required_df = required_df.sort_values(by = 'Frequency', ascending=[False])
```

```python
required_df
```

Storing frequency and rank of ingredients

```python
arr_freq = []
arr_rank = []

total_freq = 0
for item in required_df['Frequency']:
  total_freq += item

curr_idx = 1
for index, row in required_df.iterrows():
  arr_freq.append((required_df['Frequency'][index]/total_freq))
  arr_rank.append((curr_idx))
  curr_idx += 1
```

#Correct Freq-Rank Distribution

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15,10))
plt.xscale("log")
plt.yscale("log")
plt.title('Normalized Freq-Rank Distribution Plot (on log scale)')
plt.xlabel('Rank of Ingredient')
plt.ylabel('Normalized Frequency')
plt.plot(arr_rank, arr_freq, '-k')
plt.savefig('Q3.svg')
plt.savefig('Q3.png')
plt.show()
```
